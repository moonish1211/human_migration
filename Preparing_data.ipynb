# Preparind Subset of the data to different csv file

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
## Reading in data
big_data = pd.read_csv('data/Global_Human_Migration.csv')

##Since you are on Mac, the directory may be different, correct the code if it's wrong
# big_data = pd.read_csv('data\Global_Human_Migration.csv')
```

```python
big_data.columns = ['Origin_name', 'Origin_code', 'Gender_Name', 'Gender_Code', 'Destination_name', "Destination_code", '1960', '1970', '1980', '1990', '2000']
big_data = big_data.drop('Gender_Name', axis=1)
```
